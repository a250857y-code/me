# Earthquake Impact Visualization
USGSã®ãƒªã‚¢ãƒ«ã‚¿ã‚¤ãƒ åœ°éœ‡ãƒ‡ãƒ¼ã‚¿ã‚’ä½¿ã£ã¦ã€ä¸–ç•Œä¸­ã®åœ°éœ‡åˆ†å¸ƒã‚’å¯è¦–åŒ–ã—ã¾ã—ãŸã€‚
åœ°çƒã®æ´»å‹•ã‚’ç›´æ„Ÿçš„ã«æ„Ÿã˜ã‚‰ã‚Œã‚‹ã‚ˆã†ã€ãƒã‚°ãƒ‹ãƒãƒ¥ãƒ¼ãƒ‰ã”ã¨ã«è‰²åˆ†ã‘ã—ã¦ã„ã¾ã™ã€‚

```python
import pandas as pd
import folium

# USGSåœ°éœ‡ãƒ‡ãƒ¼ã‚¿ï¼ˆéå»1ã‹æœˆï¼‰
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv"
df = pd.read_csv(url)

# æ—¥æœ¬æ™‚é–“ã«å¤‰æ›
df['time'] = pd.to_datetime(df['time'], utc=True).dt.tz_convert('Asia/Tokyo')

# ğŸŒ åœ°å›³ã®ä¸­å¿ƒã‚’ã€Œæ—¥æœ¬ä»˜è¿‘ã€ã«å¤‰æ›´ï¼ˆæˆæ¥­ã¨ã®å·®åˆ¥åŒ–ï¼‰
m = folium.Map(location=[35.0, 140.0], zoom_start=3, tiles="OpenStreetMap")

# ãƒã‚°ãƒ‹ãƒãƒ¥ãƒ¼ãƒ‰ã«ã‚ˆã‚‹è‰²åˆ†ã‘ï¼ˆé–¾å€¤ã‚’å¾®èª¿æ•´ï¼‰
def color_by_mag(mag):
    if mag < 3.5:
        return 'green'
    elif mag < 5.5:
        return 'orange'
    else:
        return 'red'

# ãƒãƒ¼ã‚«ãƒ¼è¿½åŠ ï¼šã‚µã‚¤ã‚ºã¨é€æ˜åº¦ã‚’èª¿æ•´ã—ã¦è¦–èªæ€§UP
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=2 + (row['mag'] * 1.5),   # â† å°‘ã—å¼·èª¿
        color=color_by_mag(row['mag']),
        fill=True,
        fill_opacity=0.6,
        tooltip=f"{row['place']}<br>"
                f"M{row['mag']:.1f} | æ·±ã•{row['depth']}km<br>"
                f"{row['time'].strftime('%Y-%m-%d %H:%M')} JST"
    ).add_to(m)

# HTMLã¨ã—ã¦ä¿å­˜
m.save("earthquake_map.html")
print("âœ… 'earthquake_map.html' ã«åœ°å›³ã‚’ä¿å­˜ã—ã¾ã—ãŸã€‚")
print("â†’ ãƒ€ãƒ–ãƒ«ã‚¯ãƒªãƒƒã‚¯ã§ãƒ–ãƒ©ã‚¦ã‚¶ã‹ã‚‰ç¢ºèªã§ãã¾ã™ã€‚")
```
